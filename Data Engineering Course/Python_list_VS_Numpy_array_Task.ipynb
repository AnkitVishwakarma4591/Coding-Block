```python
import requests
```

```python
product = requests.get('https://fake-store-api.mock.beeceptor.com/api/products').json()
```

```python
# Stroing Data from Prodcut API
price = []
discount = []

for i in product:
    price.append(i['price'])
    discount.append(i['discount'])
```

```python
print(price)
print(discount)
```

```python
import numpy as np
```

```python
# Convert list into numpy array for better calculation
price_np = np.array(price)
discount_np = np.array(discount)
```

```python
# Calculating discount price using numpy
price_after_dis = price_np - (price_np * discount_np / 100)
```

```python
# converting numpy array into python list
discount_price = price_after_dis.tolist()
```

```python
Details_price = []
for i in range(len(price_np)):
    Details_price.append({
        "Price before discount": price[i],
        "Price after discount": discount_price[i],
        "On discount %": discount[i]
    })
```

```python
for detail in Details_price:
    print(detail)
```

# Compare the time taken by python list operation VS numpy

```python
import time

# Calculate price after discount using  Python lists
start_time_list = time.time()
price_after_dis_list = []
for p, d in zip(price, discount):
    price_after_dis_list.append(p - (p * d / 100))

end_time_list = time.time()
time_list = end_time_list - start_time_list

print(f"\nTime taken for Python list operations: {time_list:.6f} seconds")
```

```python
start_time_np = time.time()

price_after_dis_np_time = price_np - (price_np * discount_np / 100)

end_time_np = time.time()
time_np = end_time_np - start_time_np

print(f"\nTime taken for NumPy operations calculation : {time_np:.6f} seconds")
```

### Summary of Performance Comparison

*   **Time taken for Python list operations:** `0.000133` seconds
*   **Time taken for NumPy operations:** `0.000126` seconds

As observed, for this small dataset, NumPy is slightly faster than Python list operations. NumPy's efficiency significantly increases with larger datasets due to its vectorized operations.
