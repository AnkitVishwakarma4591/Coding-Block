```python
no_of_oreders = [10,20,100,400,500]
k = 3
max_sum = 0

for i in range(0,k):
  max_sum += no_of_oreders[i]

peak_win_ele = no_of_oreders[0:k]

curr_sum = max_sum
for i in range(k, len(no_of_oreders)):
    curr_sum += no_of_oreders[i]
    curr_sum -= no_of_oreders[i-k]

    max_sum = max(max_sum, curr_sum)

    if curr_sum == max_sum:
        peak_win_ele = no_of_oreders[i-k+1:i+1]

print("Max order Place :",max_sum)
print("Hours Values :",peak_win_ele)
```

```python
def isPalindrome(s):
  filtered_chars = "".join(char.lower() for char in s if char.isalnum())

  return filtered_chars == filtered_chars[::-1]
```

```python
s = "A man, a plan, a canal: Panama"
print(isPalindrome(s))
```
