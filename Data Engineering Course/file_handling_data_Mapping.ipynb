```python
with open("logs_info.log", "w") as f:
  f.writelines(data)

print("logs_info.log created")
```

```python
with open("logs_info.log", "r") as f:
  info = f.readlines()
```

```python
for line in info:
  parts = line.strip().split(' ', 2)
  if len(parts) >= 3:
    date = parts[0]
    log_level_and_message = parts[2].split(' ', 1)
    log_level = log_level_and_message[0]
    message = log_level_and_message[1] if len(log_level_and_message) > 1 else ''
    print(f"{date} : {log_level} {message}")
  else:
    print(f"Skipping malformed line: {line.strip()}")
```

# Working with CSV

```python
import requests
import csv
```

```python
users = requests.get('https://fake-store-api.mock.beeceptor.com/api/users').json()
```

```python
users
```

```python
csv_data = []
```

```python
csv_data.append(['user_id', 'username', 'email', 'password'])
```

```python
csv_data
```

```python
for user in users:
  data = [user['user_id'], user['username'], user['email'], user['password']]
  csv_data.append(data)
```

```python
csv_data
```

```python
with open("users.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(csv_data)
```
