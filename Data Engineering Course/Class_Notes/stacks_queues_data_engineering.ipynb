## 1. Stack

A stack follows LIFO: Last In, First Out.

The last item added is the first item removed.

Common operations:
- push
- pop
- peek
- is_empty

## 3. Stack Operations


```python
def push(stack, item):
    stack.append(item)

def pop(stack):
    if len(stack) == 0:
        return None
    return stack.pop()

def peek(stack):
    if len(stack) == 0:
        return None
    return stack[-1]

def is_empty(stack):
    return len(stack) == 0

pipeline_stack = []

push(pipeline_stack, read_raw_file())
push(pipeline_stack, validate_schema())
push(pipeline_stack, write_to_staging())

print("Stack:", pipeline_stack)
print("Top:", peek(pipeline_stack))
print("Removed:", pop(pipeline_stack))
print("Stack after pop:", pipeline_stack)
print("Empty:", is_empty(pipeline_stack))
```

## 4. Data Engineering Use Case: Pipeline Rollback Using Stack

A data pipeline may execute multiple steps.

If a later step fails, rollback should happen in reverse order.

Stack is suitable because rollback follows LIFO.


```python
executed_steps = []

executed_steps.append("created_temp_table")
executed_steps.append("loaded_raw_data")
executed_steps.append("validated_records")
executed_steps.append("created_final_table")

print("Executed steps:", executed_steps)

print("Pipeline failed during final validation")
print("Rollback started")

while executed_steps:
    step = executed_steps.pop()
    print("Rolling back:", step)
    print(executed_steps)
```

## 9. Queue

A queue follows FIFO: First In, First Out.

The first item added is the first item removed.

Common operations:
- enqueue
- dequeue
- peek
- is_empty


## 10. Queue Using Python List

This works for small examples, but removing from the front using pop(0) is not efficient for large data.


```python
queue = []

queue.append("file_1.csv")
queue.append("file_2.csv")
queue.append("file_3.csv")

print(queue)

first_file = queue.pop(0)
print(first_file)

print(queue)
```

## 11. Queue Using collections.deque

collections.deque is preferred for queues because removing from the left is efficient.


```python
from collections import deque

file_queue = deque()

file_queue.append("file_1.csv")
file_queue.append("file_2.csv")
file_queue.append("file_3.csv")

print(file_queue)

next_file = file_queue.popleft()
print(next_file)

print(file_queue)
```

## 12. Queue Operations


```python
from collections import deque

def enqueue(queue, item):
    queue.append(item)

def dequeue(queue):
    if len(queue) == 0:
        return None
    return queue.popleft()

def peek_queue(queue):
    if len(queue) == 0:
        return None
    return queue[0]

def is_queue_empty(queue):
    return len(queue) == 0

api_queue = deque()

enqueue(api_queue, "page_1")
enqueue(api_queue, "page_2")
enqueue(api_queue, "page_3")

print("Queue:", api_queue)
print("Front:", peek_queue(api_queue))
print("Dequeued:", dequeue(api_queue))
print("Queue after dequeue:", api_queue)
print("Empty:", is_queue_empty(api_queue))
```

## 13. Data Engineering Use Case: API Page Processing Queue

When ingesting paginated API data, pages should usually be processed in order.

Queue is suitable because page processing follows FIFO.


```python
from collections import deque

page_queue = deque()

for page_number in range(1, 6):
    page_queue.append(f"https://api.example.com/orders?page={page_number}")

while page_queue:
    page_url = page_queue.popleft()
    print("Fetching:", page_url)
```

## 15. Data Engineering Use Case: Retry Queue for Failed Records

Failed records can be placed into a retry queue.

The system can retry them later without losing the original order.


```python
from collections import deque

records = [
    {"record_id": 1, "status": "valid"},
    {"record_id": 2, "status": "failed"},
    {"record_id": 3, "status": "valid"},
    {"record_id": 4, "status": "failed"}
]

retry_queue = deque()
processed_records = []

for record in records:
    if record["status"] == "valid":
        processed_records.append(record)
    else:
        retry_queue.append(record)

print("Processed records:", processed_records)
print("Retry queue:", retry_queue)

while retry_queue:
    failed_record = retry_queue.popleft()
    print("Retrying record:", failed_record["record_id"])
```

## 16. Data Engineering Use Case: Streaming Event Buffer

## 19. Stack vs Queue


```python
stack = []
stack.append("task_1")
stack.append("task_2")
stack.append("task_3")

print("Stack output:")
while stack:
    print(stack.pop())

from collections import deque

queue = deque()
queue.append("task_1")
queue.append("task_2")
queue.append("task_3")

print("Queue output:")
while queue:
    print(queue.popleft())
```

## 20. Combined Use Case: Pipeline Execution with Retry Queue and Rollback Stack

This example uses:
- stack for rollback
- queue for task execution
- queue for retries


```python
from collections import deque

tasks = deque([
    {"task_name": "extract_orders", "status": "success"},
    {"task_name": "validate_orders", "status": "success"},
    {"task_name": "load_orders", "status": "failed"},
    {"task_name": "notify_team", "status": "success"}
])

executed_stack = []
retry_queue = deque()

while tasks:
    task = tasks.popleft()
    task_name = task["task_name"]

    print("Running:", task_name)

    if task["status"] == "success":
        executed_stack.append(task_name)
    else:
        print("Failed:", task_name)
        retry_queue.append(task)
        break

print("Executed stack:", executed_stack)
print("Retry queue:", retry_queue)

print("Rollback started")
while executed_stack:
    step = executed_stack.pop()
    print("Rolling back:", step)

print("Retrying failed tasks")
while retry_queue:
    failed_task = retry_queue.popleft()
    print("Retrying:", failed_task["task_name"])
```

## 21. Combined Use Case: API Ingestion System

This example simulates:
- API endpoints waiting in a queue
- processing successful endpoints
- sending failed endpoints to retry queue


```python
from collections import deque

api_endpoints = deque([
    {"url": "/api/products", "status": "success"},
    {"url": "/api/users", "status": "success"},
    {"url": "/api/orders", "status": "failed"},
    {"url": "/api/carts", "status": "success"}
])

retry_queue = deque()
processed_endpoints = []

while api_endpoints:
    endpoint = api_endpoints.popleft()

    if endpoint["status"] == "success":
        processed_endpoints.append(endpoint["url"])
    else:
        retry_queue.append(endpoint)

print("Processed endpoints:", processed_endpoints)
print("Retry endpoints:", retry_queue)
```

# Interview Questions


## 27. What is the difference between Stack and Queue?

Stack:
- follows LIFO
- last item added is removed first
- useful for rollback, undo, recursion, bracket validation

Queue:
- follows FIFO
- first item added is removed first
- useful for task scheduling, API processing, file ingestion, event buffers


## 28. Why is deque preferred over list for queues?

Python lists can be used as queues, but removing the first item using pop(0) is inefficient.

collections.deque supports efficient insertion and removal from both ends.

Queue operation:
- list pop(0) is O(n)
- deque popleft() is O(1)


## 29. Where are stacks used in data engineering?

Stacks can be used for:
- pipeline rollback
- undo transformation history
- validating config or JSON-like payloads
- depth-first task traversal
- dependency resolution


## 30. Where are queues used in data engineering?

Queues can be used for:
- file ingestion pipelines
- API page processing
- retry systems
- event stream buffers
- task scheduling
- producer-consumer workflows


## 31. Summary

Stack:
- LIFO
- use list with append() and pop()
- useful for rollback and reverse-order processing

Queue:
- FIFO
- use collections.deque
- useful for ordered processing and ingestion buffers

Data engineering relevance:
- pipeline execution
- failure handling
- retries
- event processing
- API ingestion
- file processing
