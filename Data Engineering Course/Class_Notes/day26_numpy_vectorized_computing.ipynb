## 1. Import NumPy


````python
import numpy as np

print(np.__version__)
````

## 2. Python List vs NumPy Array

Python lists are flexible general-purpose containers.

NumPy arrays are optimized for numerical computing.


````python
python_list = [10, 20, 30, 40, 50]
numpy_array = np.array([10, 20, 30, 40, 50])

print(python_list)
print(numpy_array)

print(type(python_list))
print(type(numpy_array))
````

## 3. Python Lists Can Store Mixed Types

A Python list can store multiple data types in the same collection.


````python
mixed_list = [101, "Riya", 250.75, True]

print(mixed_list)

for value in mixed_list:
    print(value, type(value))
````

## 4. NumPy Arrays Usually Store One Data Type

NumPy arrays are most efficient when all elements share the same data type.


````python
numeric_array = np.array([101, 250, 300])
print(numeric_array)
print(numeric_array.dtype)

mixed_array = np.array([101, 250.75, True])
print(mixed_array)
print(mixed_array.dtype)
````

## 5. Memory Efficiency

A NumPy array stores numeric values compactly.

A Python list stores references to Python objects.


````python
import sys

numbers_list = list(range(1000000))
numbers_array = np.array(numbers_list)

print("Python list container size:", sys.getsizeof(numbers_list))
print("NumPy array data size:", numbers_array.nbytes)
print("NumPy dtype:", numbers_array.dtype)
````

## 6. Python List Stores Object References

The list itself stores references.

Each Python integer is a separate object.


````python
sample_list = [10, 20, 30]

print("List object size:", sys.getsizeof(sample_list))

for value in sample_list:
    print("value:", value, "object size:", sys.getsizeof(value), "id:", id(value))
````

## 7. NumPy Array Memory Details

NumPy arrays expose memory-related attributes.


````python
arr = np.array([10, 20, 30], dtype=np.int64)

print(arr)
print("dtype:", arr.dtype)
print("itemsize:", arr.itemsize)
print("size:", arr.size)
print("nbytes:", arr.nbytes)
````

## 8. Why NumPy Is Faster for Math

NumPy operations are implemented in optimized low-level code.

This reduces Python loop overhead.


````python
import time

n = 1_000_000

list_a = list(range(n))
list_b = list(range(n))

array_a = np.array(list_a)
array_b = np.array(list_b)

start = time.time()
list_result = [list_a[i] + list_b[i] for i in range(n)]
end = time.time()
print("Python list addition time:", end - start)

start = time.time()
array_result = array_a + array_b
end = time.time()
print("NumPy array addition time:", end - start)
````

## 9. Data Engineering Example: Transaction Amount Processing

Vectorized operations can process entire batches of numeric values.


````python
transaction_amounts = np.array([100.50, 250.75, 900.00, 1200.25, 50.00])

tax_rate = 0.18

tax_amounts = transaction_amounts * tax_rate
final_amounts = transaction_amounts + tax_amounts

print("transactions:", transaction_amounts)
print("tax:", tax_amounts)
print("final amounts:", final_amounts)
````

2D arrays can represent matrix-like data.

Example:

## 15. Vectorization

Vectorization means applying an operation to an entire array at once instead of writing explicit Python loops.


````python
amounts = np.array([100, 200, 300, 400])

discounted_amounts = amounts * 0.90

print(discounted_amounts)
````

## 16. Loop vs Vectorized Operation


````python
amounts_list = [100, 200, 300, 400]

discounted_loop = []

for amount in amounts_list:
    discounted_loop.append(amount * 0.90)

print(discounted_loop)

amounts_array = np.array([100, 200, 300, 400])
discounted_vectorized = amounts_array * 0.90

print(discounted_vectorized)
````

## 17. Data Engineering Example: Currency Conversion


````python
prices_usd = np.array([10.5, 25.0, 100.75, 250.0])
usd_to_inr_rate = 83.2

prices_inr = prices_usd * usd_to_inr_rate

print(prices_inr)
````

## 18. Boolean Filtering

NumPy can apply conditions to entire arrays.


````python
order_amounts = np.array([100, -50, 250, 0, 500, 750])

valid_mask = order_amounts > 0

print(valid_mask)
print(order_amounts[valid_mask])
````

## 19. Data Engineering Example: Filter Valid Transactions


````python
transaction_amounts = np.array([100.5, -20.0, 450.0, 0.0, 999.99, -10.0])

valid_transactions = transaction_amounts[transaction_amounts > 0]

print(valid_transactions)
print("valid count:", valid_transactions.size)
print("valid total:", np.sum(valid_transactions))
````

## 20. Broadcasting

Broadcasting allows NumPy to apply operations between arrays of compatible but different shapes.


````python
amounts = np.array([100, 200, 300])

service_fee = 10

final_amounts = amounts + service_fee

print(final_amounts)
````

## 21. Broadcasting Across Rows


````python
sales_matrix = np.array([
    [100, 200, 300],
    [400, 500, 600],
    [700, 800, 900]
])

daily_bonus = np.array([10, 20, 30])

adjusted_sales = sales_matrix + daily_bonus

print(adjusted_sales)
````

## 22. Broadcasting Across Columns


````python
sales_matrix = np.array([
    [100, 200, 300],
    [400, 500, 600],
    [700, 800, 900]
])

store_adjustment = np.array([[1.10], [0.95], [1.05]])

adjusted_by_store = sales_matrix * store_adjustment

print(adjusted_by_store)
````

## 23. Data Engineering Example: Store-Level Tax Rates

Rows represent stores.

Columns represent product sales.

Each store has a different tax rate.


````python
store_product_sales = np.array([
    [1000, 2000, 1500],
    [1200, 1800, 1600],
    [900, 1400, 1300]
])

store_tax_rates = np.array([[0.18], [0.12], [0.05]])

tax_values = store_product_sales * store_tax_rates
final_values = store_product_sales + tax_values

print("tax values:")
print(tax_values)

print("final values:")
print(final_values)
````

## 24. Universal Functions

Universal functions, or ufuncs, are fast element-wise NumPy functions.

Examples:
- np.add
- np.sqrt
- np.multiply
- np.abs
- np.power


````python
values = np.array([1, 4, 9, 16, 25])

print(np.sqrt(values))
print(np.add(values, 10))
print(np.multiply(values, 2))
print(np.power(values, 2))
````

## 25. More ufunc Examples


````python
errors = np.array([-5, -2, 0, 3, 8])

print("absolute:", np.abs(errors))
print("squared:", np.power(errors, 2))
print("clipped:", np.clip(errors, 0, 5))
````

## 26. Aggregations

Aggregations summarize data.


````python
daily_revenue = np.array([1000, 1500, 1200, 1800, 2000])

print("sum:", np.sum(daily_revenue))
print("mean:", np.mean(daily_revenue))
print("min:", np.min(daily_revenue))
print("max:", np.max(daily_revenue))
print("std:", np.std(daily_revenue))
````

## 27. Aggregations on 2D Arrays

For a 2D array:
- axis=0 aggregates down columns
- axis=1 aggregates across rows


````python
store_sales = np.array([
    [1000, 1200, 1300],
    [900, 1100, 1050],
    [1500, 1600, 1700]
])

print("total per day:", np.sum(store_sales, axis=0))
print("total per store:", np.sum(store_sales, axis=1))
print("average per day:", np.mean(store_sales, axis=0))
print("average per store:", np.mean(store_sales, axis=1))
````

## 28. Data Quality Checks with NumPy


````python
amounts = np.array([100, 250, -50, 0, 700, 900, -10])

invalid_mask = amounts <= 0
valid_mask = amounts > 0

print("invalid values:", amounts[invalid_mask])
print("valid values:", amounts[valid_mask])
print("invalid count:", np.sum(invalid_mask))
print("valid count:", np.sum(valid_mask))
````

## 29. Outlier Detection

A simple outlier rule can use mean and standard deviation.


````python
latency_ms = np.array([120, 130, 125, 140, 135, 500, 128, 132])

mean_latency = np.mean(latency_ms)
std_latency = np.std(latency_ms)

threshold = mean_latency + 2 * std_latency

outliers = latency_ms[latency_ms > threshold]

print("mean:", mean_latency)
print("std:", std_latency)
print("threshold:", threshold)
print("outliers:", outliers)
````

## 30. Normalize Numeric Feature

Normalization is common before analytics and machine learning.


````python
amounts = np.array([100, 200, 300, 400, 500])

min_value = np.min(amounts)
max_value = np.max(amounts)

normalized = (amounts - min_value) / (max_value - min_value)

print(normalized)
````

## 31. Missing Values with NumPy

NumPy uses np.nan for missing numeric values.


````python
values = np.array([100.0, 200.0, np.nan, 400.0])

print(values)
print(np.isnan(values))
print(np.nanmean(values))
print(np.nansum(values))
````

## 32. Replacing Missing Values


````python
values = np.array([100.0, 200.0, np.nan, 400.0])

mean_value = np.nanmean(values)

clean_values = np.where(np.isnan(values), mean_value, values)

print(clean_values)
````

# Interview Questions


## 39. Why is a NumPy array more memory-efficient than a Python list?

A Python list stores references to Python objects.

Each object has extra overhead.

A NumPy array stores values of the same type in a compact contiguous memory block.

This makes NumPy arrays more memory-efficient for numerical data.


## 40. Explain Broadcasting with an Example

Broadcasting allows NumPy to perform operations on arrays with different shapes when they are compatible.

Example:

```python
arr = np.array([100, 200, 300])
arr + 10
```

The scalar value 10 is applied to every element.

Result:

```python
[110, 210, 310]
```


## 41. What Does Vectorization Mean in Data Processing?

Vectorization means applying operations to an entire array at once instead of using explicit Python loops.

Example:

```python
amounts * 1.18
```

This applies the operation to all elements.

Benefits:
- faster computation
- cleaner code
- less Python loop overhead
- optimized low-level execution


## 42. Why Are Python Lists Slower for Math?

Python lists are general-purpose containers.

They store references to objects and support mixed data types.

Mathematical operations require Python to process each element individually.

NumPy arrays use fixed-type contiguous memory and optimized compiled operations.


## 43. What Is a ufunc?

A ufunc, or universal function, is a fast NumPy function that operates element-wise on arrays.

Examples:
- np.add
- np.sqrt
- np.multiply
- np.abs
- np.log

ufuncs are implemented efficiently and support broadcasting.


## 44. Summary

Key takeaways:
- NumPy arrays are efficient for numerical data.
- NumPy arrays store same-type values compactly.
- Vectorization avoids explicit Python loops.
- Broadcasting applies operations across compatible shapes.
- ufuncs provide fast element-wise operations.
- Aggregations summarize arrays efficiently.
- NumPy is a foundation for Pandas, machine learning, and data processing libraries.
