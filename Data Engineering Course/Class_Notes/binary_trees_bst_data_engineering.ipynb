## 1. Tree Hierarchy Terms

A tree is a hierarchical data structure.

Common terms:
- Root: top node of the tree
- Parent: node that has children
- Child: node connected below another node
- Leaf: node with no children
- Subtree: smaller tree inside a larger tree


## 2. Basic Tree Node


```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

root = TreeNode("orders")
root.left = TreeNode("completed_orders")
root.right = TreeNode("pending_orders")

print(root.value)
print(root.left.value)
print(root.right.value)
```

## 3. Binary Tree

A binary tree is a tree where each node can have at most two children:
- left child
- right child


```python
root = TreeNode("raw_data")
root.left = TreeNode("valid_records")
root.right = TreeNode("invalid_records")

root.left.left = TreeNode("high_value_orders")
root.left.right = TreeNode("normal_orders")

print(root.value)
print(root.left.value)
print(root.right.value)
print(root.left.left.value)
print(root.left.right.value)
```

## 4. Binary Search Tree

A Binary Search Tree follows this rule:

Left subtree values < Root value < Right subtree values

This rule makes searching faster than checking every value one by one.


```python
root = TreeNode(50)
root.left = TreeNode(30)
root.right = TreeNode(70)
root.left.left = TreeNode(20)
root.left.right = TreeNode(40)
root.right.left = TreeNode(60)
root.right.right = TreeNode(80)

print(root.value)
print(root.left.value, root.right.value)
```

## 5. Insert into a BST


```python
class BST:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = TreeNode(value)
            return

        current = self.root

        while True:
            if value < current.value:
                if current.left is None:
                    current.left = TreeNode(value)
                    return
                current = current.left
            else:
                if current.right is None:
                    current.right = TreeNode(value)
                    return
                current = current.right

bst = BST()

for value in [50, 30, 70, 20, 40, 60, 80]:
    bst.insert(value)

print(bst.root.value)
print(bst.root.left.value, bst.root.right.value)
```

## 6. Search in a BST

BST search uses the sorted structure.

If target is smaller than current node, go left.

If target is larger than current node, go right.


```python
def search_bst(root, target):
    current = root

    while current is not None:
        if current.value == target:
            return True
        elif target < current.value:
            current = current.left
        else:
            current = current.right

    return False

print(search_bst(bst.root, 60))
print(search_bst(bst.root, 99))
```

## 7. Data Engineering Example: Search Sorted Order IDs

A BST can represent indexed values such as order IDs, customer IDs, or partition boundaries.


```python
order_id_bst = BST()

order_ids = [1005, 1002, 1008, 1001, 1003, 1007, 1010]

for order_id in order_ids:
    order_id_bst.insert(order_id)

print(search_bst(order_id_bst.root, 1007))
print(search_bst(order_id_bst.root, 9999))
```

## 8. In-order Traversal

In-order traversal order:

Left -> Root -> Right

For a BST, in-order traversal gives sorted output.


```python
def inorder(root):
    if root is None:
        return []

    return inorder(root.left) + [root.value] + inorder(root.right)

print(inorder(bst.root))
print(inorder(order_id_bst.root))
```

## 9. Pre-order Traversal

Pre-order traversal order:

Root -> Left -> Right

Useful when copying or serializing a tree structure.


```python
def preorder(root):
    if root is None:
        return []

    return [root.value] + preorder(root.left) + preorder(root.right)

print(preorder(bst.root))
```

## 10. Post-order Traversal

Post-order traversal order:

Left -> Right -> Root

Useful when deleting or cleaning child nodes before parent nodes.


```python
def postorder(root):
    if root is None:
        return []

    return postorder(root.left) + postorder(root.right) + [root.value]

print(postorder(bst.root))
```

## 11. Traversal Data Engineering Analogy

In-order:
- useful for reading indexed values in sorted order

Pre-order:
- useful for saving structure before children

Post-order:
- useful for cleanup or deletion workflows where children should be processed before parent


## 12. Maximum Nodes at Level L

In a binary tree, the maximum number of nodes at level L is:

2^L

Assuming root is at level 0:
- Level 0: 1 node
- Level 1: 2 nodes
- Level 2: 4 nodes
- Level 3: 8 nodes


```python
def max_nodes_at_level(level):
    return 2 ** level

for level in range(5):
    print("Level:", level, "Max nodes:", max_nodes_at_level(level))
```

## 13. Search Complexity in BST

Balanced BST:
- search: O(log n)

Unbalanced BST:
- search: O(n)

A balanced BST keeps height small.

An unbalanced BST can become like a linked list.


```python
import math

record_counts = [10, 100, 1000, 1000000, 1000000000]

for n in record_counts:
    approx_steps = math.ceil(math.log2(n))
    print("Records:", n, "Approx balanced BST search steps:", approx_steps)
```

## 14. Balanced vs Unbalanced BST


```python
balanced_values = [50, 30, 70, 20, 40, 60, 80]
unbalanced_values = [10, 20, 30, 40, 50, 60, 70]

balanced_bst = BST()
unbalanced_bst = BST()

for value in balanced_values:
    balanced_bst.insert(value)

for value in unbalanced_values:
    unbalanced_bst.insert(value)

print("Balanced BST inorder:", inorder(balanced_bst.root))
print("Unbalanced BST inorder:", inorder(unbalanced_bst.root))
```

## 15. Data Engineering Context: Database Indexes

Databases use tree-like index structures to avoid scanning every row.

Instead of checking every row:
- index helps narrow down the search
- fewer blocks/pages are read
- queries become faster

Common database indexes are often based on B-Tree or B+Tree structures, not simple binary trees.

The core idea is similar:
- tree structure helps skip unnecessary data


## 16. Data Engineering Context: Parquet and ORC Data Skipping

Columnar file formats such as Parquet and ORC store metadata.

This metadata can include:
- min value
- max value
- row group statistics
- column-level statistics

Query engines can skip chunks of data that cannot contain the target value.

This is conceptually similar to using index/tree-like metadata for faster retrieval.


```python
row_groups = [
    {"file": "part_1.parquet", "min_order_id": 1000, "max_order_id": 1999},
    {"file": "part_2.parquet", "min_order_id": 2000, "max_order_id": 2999},
    {"file": "part_3.parquet", "min_order_id": 3000, "max_order_id": 3999},
]

target_order_id = 2500

files_to_scan = []

for group in row_groups:
    if group["min_order_id"] <= target_order_id <= group["max_order_id"]:
        files_to_scan.append(group["file"])

print(files_to_scan)
```

# Interview Questions


## 20. What is the maximum number of nodes at level L in a binary tree?

The maximum number of nodes at level L is:

2^L

If root is level 0:
- level 0 has 1 node
- level 1 has 2 nodes
- level 2 has 4 nodes
- level 3 has 8 nodes


## 21. How does a Binary Search Tree speed up data retrieval?

A BST speeds up retrieval by reducing the search space at every step.

At each node:
- if target is smaller, search left
- if target is larger, search right
- if equal, value is found

In a balanced BST, this avoids scanning every node.


## 22. What is the time complexity of searching in a balanced BST?

Searching in a balanced BST takes:

O(log n)

because the search space is approximately divided in half at each level.

For an unbalanced BST, search can degrade to:

O(n)
